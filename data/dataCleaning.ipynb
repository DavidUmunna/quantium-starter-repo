```python
# concatenate the three dataframes vertically into a single dataframe
Combined_Sales_df = pd.concat([data_df_0, data_df_1, data_df_2], ignore_index=True)
```

```python
Combined_Sales_df.head()
```

```python
Combined_Sales_df.to_csv("Combined_Sales_df.csv")
```

```python
# convert price strings like '$3.00' to float 3.00
Combined_Sales_df['price'] = Combined_Sales_df['price'].str.replace(r'[\$,]', '', regex=True).astype(float)
```
